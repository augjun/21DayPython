# 绘制折线图

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
# 绘制多组数据
import numpy as np
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

# 绘制散点图

```python
import numpy as np
x = np.arange(20)
noise = np.random.randn(20)
y = 5*x + 5 + noise *10
plt.scatter(x, y)
plt.show()
```

# 绘制直方图

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
plt.hist(x, 50)
plt.show()
```

# 绘制柱状图

```python
models = ['iphone11', 'mate30', 'Xiaomi9']
sales = [20, 30, 15]
plt.bar(models, sales)
plt.show()
```

# 绘制饼图

```python
models = ['iphone11', 'iphoneX', 'iphoneXs']
sales = [30, 20, 10]
plt.pie(sales, explode=[0,0,0.1], labels=models, autopct='%.2f%%')
plt.show()
```

# 常用的可定制属性

```python
# 设置颜色
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.arange(10)
y = np.arange(10) ** 2

ax.plot(x, y, color='g')
ax.plot(x, y, color='#CECECE')
ax.scatter(x, y, color='r')
```

```python
# 设置标记
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(x, y, marker='o')
ax.plot(x, y, marker='*')
```

```python
# 设置线型
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(x, y, linestyle='-')
ax.plot(x, y, linestyle='--')
```

```python
# 设置刻度位置
x = np.arange(100)
y = np.sin(x/50*np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.set_xticks([0, 50, 100])
ax.set_yticks([-1, 0, 1])
ax.plot(x, y)
```

```python
# 设置刻度标签
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.arange(100)
y = np.sin(x/50*np.pi)
ax.set_xticks([0, 50, 100])

labels = ax.set_xticklabels(['one','two','three'])
ax.plot(x,y)
```

```python
# 设置图表范围
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.arange(100)
y = np.sin(x/50*np.pi)

ax.set_xlim(0, 200)
ax.set_ylim(-10, 10)
ax.plot(x,y)
```

```python
# 设置标题
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.arange(100)
y = np.sin(x/50*np.pi)

ax.set_title('Sine Wave')
ax.plot(x,y)
```

```python
# 设置轴标签
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.arange(100)
y = np.sin(x/50*np.pi)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(x,y)
```

```python
# 批量设定属性
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.arange(100)
y = np.sin(x/50*np.pi)

props = {
    'title': 'Sine Wave',
    'xlabel': 'x',
    'ylabel': 'y',
}
ax.set(**props)
ax.plot(x,y)
```

```python
# 添加图例
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.arange(100)
y = np.sin(x/50*np.pi)
y2 = np.sin(x/50*np.pi) + 2

ax.plot(x, y, label='one')
ax.plot(x, y2, label='two')
ax.legend()
```

```python
# 添加注解
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.arange(100)
y = np.sin(x/50*np.pi)
ax.text(50, 1, 'Hello world!',
  family='monospace', fontsize=10)
ax.plot(x, y)
```

```python
# 添加指向性注解
x = np.arange(100)
y = np.sin(x/50*np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xticks([0, 50, 100])
ax.set_yticks([-1, 0, 1])
ax.set_ylim(-1.5, 1.5)

ax.annotate('peak', xy=(25,1),
                xytext=(30, 1.5),
                arrowprops=dict(facecolor='black', headwidth=6, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')
ax.annotate('valley', xy=(75,-1),
                xytext=(80, -0.5),
                arrowprops=dict(facecolor='black', headwidth=6, width=2,
                                headlength=4),
                horizontalalignment='left', verticalalignment='top')
ax.plot(x,y, label='one')
ax.legend()
```

```python
# 自动调整x轴上的日期，为了更清楚的显示
import pandas as pd
dates=pd.date_range('2019-01-01', periods=100)
data = np.random.randn(100).cumsum()
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.gcf().autofmt_xdate()
ax.plot(dates, data)
```

# 实战项目 - 见PPT

```python
# 调整图像显示大小
import pylab
pylab.rcParams['figure.figsize'] = (15.0, 8.0)
```

```python
import pandas as pd
df = pd.read_csv("houses.csv", sep='>', index_col=0)
df = df.drop_duplicates(["title", "price",
                    "xiaoqu_name", "size",
                    "buildyear", "huxing",
                    "chaoxiang", "zhuangxiu",
                    "cenggao", "district_name",
                    "sub_district_name"])
df.head(2)
```

```python
# 每个区的均价
df_avg = df.groupby("district_name")["up_price"].agg(avg_price="mean").sort_index()
df_avg["district_pinyin"] = ["jiading", "fengxian", "baoshan", "xuhui", "putuo", "yangpu", "songjiang", "pudong",
                            "hongkou", "jinshan", "changning", "minhang", "qingpu", "jingan", "huangpu"]
df_avg.sort_values("avg_price", inplace=True)
plt.bar(df_avg["district_pinyin"], df_avg["avg_price"])
```

```python
# 房源面积的分布情况
plt.hist(df["size"], bins=100)
```

```python
# 房源在各个区的分布情况
df_district = df.groupby("district_name")["price"].agg(house_num="count").sort_index()
df_district["district_pinyin"] = ["jiading", "fengxian", "baoshan", "xuhui", "putuo", "yangpu", "songjiang", "pudong",
                            "hongkou", "jinshan", "changning", "minhang", "qingpu", "jingan", "huangpu"]
df_district.sort_values("house_num", inplace=True)
plt.pie(df_district["house_num"], labels=df_district["district_pinyin"])
```

```python
# 房源的装修情况
df_zhuangxiu = df[df.zhuangxiu!=""].groupby("zhuangxiu")["price"].agg(house_num="count").sort_values(by="house_num")
df_zhuangxiu["zhuangxiu_pinyin"] = ["maopie", "qita", "jianzhuang", "jingzhuang"]
df_zhuangxiu.sort_values("house_num", inplace=True)
plt.pie(df_zhuangxiu["house_num"], labels=df_zhuangxiu["zhuangxiu_pinyin"])
```

```python
# 房价和面积的关系
plt.scatter(df["size"], df["price"], alpha=0.3)
```

```python
# 选取一个小区的房源，查看房价和面积的关系
df_xiaoqu = df.groupby("xiaoqu_name")["price"].agg(house_num="count").sort_values(by="house_num")
df_xiaoqu.tail(1)
```
